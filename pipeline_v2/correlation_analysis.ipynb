```python
import pandas as pd
```

```python
def highlight_attributes(att):
    att = att.apply(lambda x: abs(x))
    sorted_att = att.sort_values(ascending=True)
    sorted_list = list(sorted_att[:4])
    return ['background-color: grey' if value in sorted_list else '' for value in att]
```

```python
df = pd.read_csv("create_database/df_database.csv", index_col=0)
df = df.corr()
final_table = df.style.apply(highlight_attributes)
final_table
```
